考虑到支付客流的增长趋势，这里只考虑2016.8.1之后的数据，并且剔除国庆和中秋假期

```python
data_pay[(True^data_pay.time_stamp.isin(pd.date_range('2016-10-01', '2016-10-10'))) & (True^data_pay.time_stamp.isin(pd.date_range('2016-09-12', '2016-09-19'))) & (data_pay.time_stamp > pd.to_datetime('2016-08-01'))]
```

```python
feature_daypay=pd.DataFrame({'daypay': data_pay.groupby(['shop_id','date'])['user_id'].count()}).reset_index()
```

```python
# feature_daypay.head()
feature_daypay['date'] = pd.to_datetime(feature_daypay.date, format='%Y-%m-%d')
```

```python
feature_daypay['weekday'] = feature_daypay.date.dt.weekday
```

```python
weekday_sum = feature_daypay.groupby(['shop_id','weekday'])['daypay'].sum()
weekday_count = feature_daypay.groupby(['shop_id','weekday'])['daypay'].count()
weekday_mean = pd.DataFrame({'average': weekday_sum / weekday_count}).reset_index()
weekday_mean[:50]
```

```python
# weekday_mean
tmp = weekday_mean.pivot(index='shop_id',columns='weekday',values='average').reset_index()
predict_mean = pd.merge(tmp, tmp.copy(), on='shop_id')
tmp_columns = ['day_'+str(i) for i in range(14)]
tmp_columns.insert(0, 'shop_id')
predict_mean.columns = tmp_columns
# predict_mean
```

```python
predict_mean.round(0)
predict_mean = predict_mean.astype(int)
```

```python
predict_mean.to_csv('../tempfile/predict_mean.csv', index=False, header=False)
```
